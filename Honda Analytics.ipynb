# Data Cleaning
We can now read the csv file we created above!

```python
df2 = pd.read_csv("honda_sell_data.csv")
```

```python
df2.head(3)
```

Turning all '-' into NaN

```python
cols = ['Year','Make','Model','Condition','Price','Consumer_Rating','Consumer_Review_#','Exterior_Color','Interior_Color','Drivetrain','MPG','Fuel_Type','Transmission','Engine','VIN','Stock_#','Mileage','Comfort_Rating','Interior_Design_Rating','Performance_Rating','Value_For_Money_Rating','Exterior_Styling_Rating','Reliability_Rating', 'State', 'Seller_Type']
df2[cols] = df2[cols].replace({"–": np.nan})
```

- We have to make Price a Integer
- We have to make Mileage an Float

```python
df2["Price"] = df2["Price"].str.replace("$", "", regex=False).str.replace(",", "", regex=False).str.replace("Not Priced", "0", regex=False).astype(float)
df2["Price"] = df2["Price"].replace(0.0, np.nan, regex=False)
df2["Mileage"] = df2["Mileage"].str.replace("USB", "100.1234", regex=False).str.replace("Premium", "100.1234", regex=False).str.replace("HomeLinkRear", "100.1234", regex=False).str.replace("Alloy", "100.1234", regex=False).str.replace("Bluetooth100.1234", "100.1234", regex=False).str.replace("Apple", "100.1234", regex=False).astype(float)
df2["Mileage"] = df2["Mileage"].replace(100.1234, np.nan, regex=False)

```

```python
# Replaces the null values from Prices with it's models average
def average_price():
    average_price_per_model = df2.groupby("Model")[["Price"]].agg(np.mean)
    null_prices_rows = df2[df2["Price"].isna()][["Model","Price"]]
    for index in list(null_prices_rows.index):
        model = null_prices_rows.loc[[index]]['Model'].item()
        df2.at[index,'Price'] = average_price_per_model.loc[[model]]['Price'].item()
average_price()
```

```python
def median_ratings():
    for rating_type in ['Comfort_Rating','Interior_Design_Rating','Performance_Rating','Value_For_Money_Rating','Exterior_Styling_Rating','Reliability_Rating']:
        average_rating_type_car_model = df2[[rating_type,"Model"]].groupby("Model").agg(np.median)
        null_rating_type_rows = df2[df2[rating_type].isna()][["Model",rating_type]]
        for index in list(null_rating_type_rows.index):
            model = null_rating_type_rows.loc[[index]]['Model'].item()
            df2.at[index,rating_type] = average_rating_type_car_model.loc[[model]][rating_type].item()
median_ratings()
```

```python

df3 = df2.dropna(subset=['Comfort_Rating','State','Seller_Type','Transmission','Exterior_Color','Drivetrain','Mileage']).drop(['Interior_Color'], axis=1).drop_duplicates()

df3['min_MPG'] = df3['MPG'].str.split("–",expand = True)[0]
df3['max_MPG'] = df3['MPG'].str.split("–",expand = True)[1]
df3.drop(['MPG'], axis=1, inplace=True)

df3[['min_MPG', 'max_MPG']] = df3[['min_MPG', 'max_MPG']].fillna(0)
df3['min_MPG'] = df3['min_MPG'].astype(float)
df3['max_MPG'] = df3['max_MPG'].astype(float)
df3['avg_MPG'] = (df3['min_MPG']+df3['max_MPG'])/2
df3[['min_MPG', 'max_MPG', 'avg_MPG']] = df3[['min_MPG', 'max_MPG', 'avg_MPG']].replace(0.0, np.nan)

median_ratings()

df3.dropna(subset=['min_MPG','max_MPG','avg_MPG'], inplace=True)
df3.drop(df3[(df3['State']=='MO-22') | (df3['State']=='Route') | (df3['State']=='Glens')].index, inplace=True)

df3['Transmission'] = df3['Transmission'].str.lower()
df3.loc[df3['Transmission'].str.contains('automatic|a/t|9-speed', regex=True), 'Transmission'] = 'automatic'
df3.loc[df3['Transmission'].str.contains('manual|m/t', regex=True), 'Transmission'] = 'manual'
df3.loc[df3['Transmission'].str.contains('cvt|variable|driver selectable mode', regex=True), 'Transmission'] = 'cvt'
df3.loc[(df3['Transmission'].str.contains('other')) & (df3['Model'].str.contains('CR-V')), 'Transmission'] = 'cvt'
df3.loc[(df3['Transmission'].str.contains('other')), 'Transmission'] = 'automatic'

df3['Drivetrain'] = df3['Drivetrain'].str.lower()
df3.loc[df3['Drivetrain'].str.contains('front-wheel drive|fwd', regex=True), 'Drivetrain'] = 'fwd'
df3.loc[df3['Drivetrain'].str.contains('all-wheel drive|awd|four-wheel drive', regex=True), 'Drivetrain'] = 'awd'
df3.loc[df3['Drivetrain'].str.contains('rear-wheel drive', regex=True), 'Drivetrain'] = 'rwd'

print(df3.shape)
df3.isna().sum()
```
